```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample hypertension dataset
np.random.seed(42)
size = 500  # Sample size

data = pd.DataFrame({
    'Cholesterol': np.random.normal(200, 30, size),  # Simulated cholesterol values
    'Hypertension': np.random.choice([0, 1], size=size, p=[0.6, 0.4])  # 0 = Low, 1 = High
})

# Create figure and axes for plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter Plot: Cholesterol vs Hypertension
sns.scatterplot(ax=axes[0], x=data['Cholesterol'], y=data['Hypertension'], alpha=0.3, color='blue')
axes[0].set_xlabel("Cholesterol Level")
axes[0].set_ylabel("Hypertension (0 = Low, 1 = High)")
axes[0].set_title("Cholesterol vs Hypertension Scatter Plot")

# Box Plot: Cholesterol distribution by Hypertension
sns.boxplot(ax=axes[1], x=data['Hypertension'], y=data['Cholesterol'], palette='coolwarm')
axes[1].set_xlabel("Hypertension (0 = Low, 1 = High)")
axes[1].set_ylabel("Cholesterol Level")
axes[1].set_title("Cholesterol Distribution by Hypertension")

# Show plots
plt.tight_layout()
plt.show()
```

```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

class HypertensionCalculator:
    def __init__(self, root, model, scaler):
        """Initialize the GUI and integrate the trained model."""
        self.root = root
        self.model = model
        self.scaler = scaler

        self.root.title("Risk of Hypertension Calculator")

        # Menu Frame
        self.menu_frame = tk.Frame(root, bg="grey", height=50, width=200)
        self.menu_frame.pack(fill=tk.X)

        # Hamburger Menu Icon
        self.hamburger_icon = tk.Label(self.menu_frame, text="☰", font=("Arial", 18), bg="grey", fg="white")
        self.hamburger_icon.pack(side=tk.LEFT, padx=10, pady=5)

        # Title Label
        self.title_label = tk.Label(root, text="Risk of Hypertension Calculator", font=("Arial", 16))
        self.title_label.pack(pady=10)

        # Input Fields
        self.create_input_fields()

        # Calculate Button
        self.calculate_button = tk.Button(root, text="Calculate", font=("Arial", 14), bg="black", fg="white",
                                          command=self.calculate_hypertension)
        self.calculate_button.pack(pady=10)

        # Prediction Label
        self.result_label = tk.Label(root, text="", font=("Arial", 14), fg="red")
        self.result_label.pack(pady=10)

    def create_input_fields(self):
        """Create input fields for user data."""
        self.input_frame = tk.Frame(self.root)
        self.input_frame.pack(pady=10)

        # Labels and Entry Fields for user inputs
        self.labels = ["Age", "BMI", "Cholesterol", "Systolic_BP", "Diastolic_BP", 
                       "Smoking_Status (0=Never, 1=Former, 2=Current)", 
                       "Alcohol_Intake", "Physical_Activity_Level (0=Low, 1=Moderate, 2=High)", 
                       "Family_History (0=No, 1=Yes)", "Glucose"]
        self.entries = {}

        for label in self.labels:
            frame = tk.Frame(self.input_frame)
            frame.pack(fill=tk.X, pady=5)

            tk.Label(frame, text=label, font=("Arial", 12)).pack(side=tk.LEFT, padx=10)
            entry = tk.Entry(frame, font=("Arial", 12))
            entry.pack(side=tk.RIGHT, padx=10)
            self.entries[label] = entry  # Store entry fields in a dictionary

    def calculate_hypertension(self):
        """Predict hypertension risk based on user inputs using ML model."""
        try:
            # Collect inputs
            user_input = np.array([float(self.entries[label].get()) for label in self.labels]).reshape(1, -1)

            # Standardize input using trained scaler
            user_input_scaled = self.scaler.transform(user_input)

            # Predict using trained model
            prediction = self.model.predict(user_input_scaled)[0]

            # Display result
            result = "High Risk of Hypertension" if prediction == 1 else "Low Risk of Hypertension"
            color = "red" if prediction == 1 else "green"
            self.result_label.config(text=result, fg=color)

        except ValueError:
            messagebox.showerror("Input Error", "Please enter valid numerical values.")

# Load trained model (from previous training)
model = LogisticRegression(class_weight="balanced", max_iter=500)
scaler = StandardScaler()

# Run Tkinter application
if __name__ == "__main__":
    root = tk.Tk()
    app = HypertensionCalculator(root, model, scaler)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import numpy as np

class HypertensionCalculator:
    def __init__(self, root):
        """Initialize the GUI and set up the prediction system."""
        self.root = root
        self.root.title("Risk of Hypertension Calculator")

        # Menu Frame
        self.menu_frame = tk.Frame(root, bg="grey", height=50, width=200)
        self.menu_frame.pack(fill=tk.X)

        # Hamburger Menu Icon
        self.hamburger_icon = tk.Label(self.menu_frame, text="☰", font=("Arial", 18), bg="grey", fg="white")
        self.hamburger_icon.pack(side=tk.LEFT, padx=10, pady=5)

        # Title Label
        self.title_label = tk.Label(root, text="Risk of Hypertension Calculator", font=("Arial", 16))
        self.title_label.pack(pady=10)

        # Input Fields (UPDATED: Removed Smoking_Status, Physical_Activity_Level, and Family_History)
        self.create_input_fields()

        # Calculate Button
        self.calculate_button = tk.Button(root, text="Calculate", font=("Arial", 14), bg="black", fg="white",
                                          command=self.calculate_hypertension)
        self.calculate_button.pack(pady=10)

        # Prediction Label
        self.result_label = tk.Label(root, text="", font=("Arial", 14), fg="red")
        self.result_label.pack(pady=10)

    def create_input_fields(self):
        """Create input fields for user data."""
        self.input_frame = tk.Frame(self.root)
        self.input_frame.pack(pady=10)

        # Labels and Entry Fields for user inputs (Removed Smoking_Status, Physical_Activity_Level, Family_History)
        self.labels = ["Age", "Cholesterol Level", "BMI"]
        self.entries = {}

        for label in self.labels:
            frame = tk.Frame(self.input_frame)
            frame.pack(fill=tk.X, pady=5)

            tk.Label(frame, text=label, font=("Arial", 12)).pack(side=tk.LEFT, padx=10)
            entry = tk.Entry(frame, font=("Arial", 12))
            entry.pack(side=tk.RIGHT, padx=10)
            self.entries[label] = entry  # Store entry fields in a dictionary

    def calculate_hypertension(self):
        """Perform risk prediction based on input values."""
        try:
            # Collect inputs
            age = float(self.entries["Age"].get())
            cholesterol = float(self.entries["Cholesterol Level"].get())
            bmi = float(self.entries["BMI"].get())

            # Simple rule-based prediction
            risk_score = (cholesterol / 200) + (bmi / 25) + (age / 50)

            # Decision threshold
            if risk_score > 2.5:
                result = "High Risk of Hypertension"
                color = "red"
            else:
                result = "Low Risk of Hypertension"
                color = "green"

            # Display Result
            self.result_label.config(text=result, fg=color)

        except ValueError:
            messagebox.showerror("Input Error", "Please enter valid numerical values.")

# Run Application
if __name__ == "__main__":
    root = tk.Tk()
    app = HypertensionCalculator(root)
    root.mainloop()
```
