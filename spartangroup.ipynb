```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample hypertension dataset
np.random.seed(42)
size = 500  # Sample size

data = pd.DataFrame({
    'Cholesterol': np.random.normal(200, 30, size),  # Simulated cholesterol values
    'Hypertension': np.random.choice([0, 1], size=size, p=[0.6, 0.4])  # 0 = Low, 1 = High
})

# Create figure and axes for plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter Plot: Cholesterol vs Hypertension
sns.scatterplot(ax=axes[0], x=data['Cholesterol'], y=data['Hypertension'], alpha=0.3, color='blue')
axes[0].set_xlabel("Cholesterol Level")
axes[0].set_ylabel("Hypertension (0 = Low, 1 = High)")
axes[0].set_title("Cholesterol vs Hypertension Scatter Plot")

# Box Plot: Cholesterol distribution by Hypertension
sns.boxplot(ax=axes[1], x=data['Hypertension'], y=data['Cholesterol'], palette='coolwarm')
axes[1].set_xlabel("Hypertension (0 = Low, 1 = High)")
axes[1].set_ylabel("Cholesterol Level")
axes[1].set_title("Cholesterol Distribution by Hypertension")

# Show plots
plt.tight_layout()
plt.show()
```

```python
# Create figure and axes for plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter Plot: Cholesterol vs Hypertension
sns.scatterplot(ax=axes[0], x=data['Cholesterol'], y=data['Hypertension'], alpha=0.3, color='blue')
axes[0].set_xlabel("Cholesterol Level")
axes[0].set_ylabel("Hypertension (0 = Low, 1 = High)")
axes[0].set_title("Cholesterol vs Hypertension Scatter Plot")

# Strip Plot: Cholesterol distribution by Hypertension
sns.stripplot(ax=axes[1], x=data['Hypertension'], y=data['Cholesterol'], jitter=True, palette='coolwarm', alpha=0.6)
axes[1].set_xlabel("Hypertension (0 = Low, 1 = High)")
axes[1].set_ylabel("Cholesterol Level")
axes[1].set_title("Cholesterol Distribution by Hypertension (Strip Plot)")

# Show plots
plt.tight_layout()
plt.show()
```

Scatter Plot (Left - Cholesterol vs Hypertension)
The majority of points are clustered at Hypertension = 0 (low hypertension).
Some points are at Hypertension = 1 (high hypertension), but they are also spread across different Cholesterol levels.
No strong visible correlation between cholesterol and hypertension (i.e., high cholesterol does not always mean high hypertension).
Strip Plot (Right - Cholesterol Distribution by Hypertension)
Both hypertension groups (0 and 1) show a similar spread of cholesterol levels.
There is some overlap between the cholesterol values of hypertensive and non-hypertensive individuals.
Hypertensive individuals (1) appear to have a slightly higher median cholesterol level, but there is no clear cut-off.
Outliers exist in both groups, suggesting variability.
Key Takeaways:
✅ Cholesterol alone is not a perfect predictor of hypertension.
✅ There is some trend where higher cholesterol might be linked to hypertension, but it's not definitive.
✅ Additional features (like BMI, smoking status, physical activity, etc.) should be included to improve predictive accuracy.
